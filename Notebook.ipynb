# Práctica 6. *Indexación y selección de datos*

```python
import pandas as pd
```

1. **Acceso a columnas y estadísticas básicas.

```python
vuelos=pd.read_csv('Data/ventas.csv',encoding='latin_1')
```

   - Imprime las primeras 5 filas del DataFrame.

```python
vuelos.head(5)
```

 - Calcula la media y la desviación estándar de la columna `num_passengers`. (Aplicar método [`mean()` y `std()`](docs/dispersión.md))

```python
# Calcular la media de las ventas en 2022
media_vuelos = vuelos['num_passengers'].mean()
print(f"Media de pasajeros: {media_vuelos}")
```

```python
# Calcular la desviación estándar de las ventas en 2023
desviacion_vuelos = vuelos['num_passengers'].std()
print(f"Desviación estándar de vuelos: {desviacion_vuelos}")
```

- ¿Cuál es el número máximo de pasajeros registrados? (Aplicar método  [`max()`](docs/docsymin.md)).

```python
# Calcular el máximo de las ventas en 2022
max_pasageros = vuelos['num_passengers'].max()
print(f"Máximo de pasageros: {max_pasageros}")
```

4. **Agregación y agrupación**:

 - Agrupa los datos por `ruta` y calcula la suma total de pasajeros para cada ruta. (aplicar  [`groupby()`, `sum()`](docs/groupbysum.md)).

```python
Ruta=vuelos.groupby('route')['num_passengers'].sum().sort_values(ascending=True)
```

 ¿Cuál es la ruta más popular en términos de pasajeros?

```python
Ruta.tail(1)
```

5. **Creación de nuevas columnas**:

6. **Visualización de datos**:

- Crea un gráfico de barras que muestre la cantidad de pasajeros por día de la semana (`flight_day`).

```python
vuelos['booking_lead_days'].mean()
```

```python
Demanda_semanal=vuelos.groupby('flight_day')['num_passengers'].mean()
```

```python
import matplotlib.pyplot as plt
```

```python
Demanda_semanal.plot.bar(x=Demanda_semanal.index,y=Demanda_semanal,rot=0)
plt.title('Demanda por dia de la semana')
plt.xlabel('Dia de la semana')
plt.ylabel('NUmeros de pasageros promedio')
plt.show()
```
